# Association: Apriori Algorithm; 1ère Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
           ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
           ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
           ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
           ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
dataset
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
apriori(df, min_support=0.5)
```

```python
apriori(df, min_support=0.5, use_colnames=True)
```

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
rules[(rules['lift'] > 1)]
rules
```

```python
rules[ (rules['antecedent_len'] >= 1) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1) ]
```

# Association: Apriori Algorithm; 2ère Dataset

```python
from apyori import apriori
```

```python
df=pd.read_csv('Market.csv',header=None)
df.head(10)
```

```python
df.columns
```

```python
liste = []  
for i in range(0, 7501):  
    liste.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
te = TransactionEncoder()
te_ary = te.fit(liste).transform(liste)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
association_rules = apriori(liste, min_support=0.0045,min_lift=3)  
association_results = list(association_rules)  
```

```python
association_results
```

```python
for item in association_results:

    
    pair = item[0]
    print(pair)
    items = [x for x in pair]
    
    print("Rule: " + items[0] + " -> " + items[1])


    print("Support: " + str(item[1]))


    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
